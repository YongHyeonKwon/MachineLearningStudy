```python
# 필요한 도구 불러오기 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 모델 불러오기 
from sklearn.neighbors import KNeighborsClassifier # 분류 모델
# 모델 성능 평가 지표 (평가 도구)
from sklearn.metrics import accuracy_score # 정확도 평가 도구 
```

- 데이터 전처리
    - 데이터 분리

```python
data
```

```python
X = data.loc[:,['Height','Weight']] # 문제('키','몸무게') 
# data.loc[:,'Height':'Weight']
# data.iloc[:,[1,2]]
# data.iloc[:,1:3] # 2까지 포함
y = data['Label']# 답 ('라벨')
# data.loc[:,'Label']
# data.iloc[:,-1] , data.iloc[:,3]
print('문제 데이터:', X.shape ) # 2차원(행:500,열:2)
print('답 데이터:', y.shape) # 1차원(데이터 개수: 500)
# 문제 500개면 답 500
```

```python
# 훈련세트와 테스트 세트로 행 분리 
# 머신러닝 훈련데이터셋
# 학습이 잘 됐는지 확인하기 위한 테스트데이터셋
# 훈련:테스트 -> 7:3
# 0~349(처음부터 350번째까지): train
# 350~(350번부터 끝까지 150개): test
# X - train, test 분리 / y - train, test 분리 
X_train= X.iloc[:350]
X_test= X.iloc[350:]

y_train= y.iloc[:350]
y_test= y.iloc[350:]

print('훈련 문제:', X_train.shape)
print('훈련 답:', y_train.shape)
print('테스트 문제:', X_test.shape)
print('테스트 답:', y_test.shape)
```

#### 모델링 
 - 모델 객체 생성 
 - 모델 학습
 - 모델 평가 및 예측

```python
# knn 분류 모델 
knn_model = KNeighborsClassifier(n_neighbors=3)
# 기본값으로 이웃의 수를 5로 설정 
# 거리가 가장 가까운 이웃 5개의 실제 답을 보면서 예측하는 모델 
```

```python
# 모델 학습
# 학습 중요 x
# 어떤 데이터를 보면서 학습하는지가 중요 o
# 학습하기 위한 좋은 재료가 필요 
# 데이터에 맞춰서 학습하는 것이 머신러닝 모델 
# 모델.fit(X_훈련,y_훈련)
knn_model.fit(X_train,y_train)
```

```python
# 모델 예측
# 모델.predict(테스트문제)
pre = knn_model.predict(X_test)
pre
# 예측이 다 맞을까? 예측은 오류가 있을 수 있음
# 일반적으로 잘 예측할 수 있게 모델을 만들어야하는 상황 
```

```python
# 실제값과 pre(예측값)이 어느정도 맞아떨어지는지 확인 
# over - over 맞았는지 확인
# 정확도 평가 지표 
# accuracy_score(실제답, 예측값)
# pre , y_test
accuracy_score(y_test,pre) * 100 # 90.7%
# 정확도 0~1 : 1에 가까울수록 예측이 잘 된 상황
```
